## Section 1 — Reading Datasets

**Goal of this section:** load a dataset that ships with scikit-learn and turn it into a `DataFrame`.

#### The Iris dataset
The **Iris dataset** is one of the most-used datasets in machine learning. It measures **150 iris flowers** — 50 of each of 3 species.

| | |
|---|---|
| Rows (flowers) | 150 |
| Features | 4 numeric measurements, in centimetres |
| Species | `setosa`, `versicolor`, `virginica` (50 each) |

**The 4 measurements:** `sepal length (cm)`, `sepal width (cm)`, `petal length (cm)`, `petal width (cm)`.

A **petal** is an inner flower leaf; a **sepal** is the outer leaf that protects the bud. The figure below shows both on all three species — run the next cell to see it.

````
# This figure lives in the lab's GitHub repo. Download it once into this Colab
# session, then open it as a local file.
import urllib.request
from IPython.display import Image

URL = "https://raw.githubusercontent.com/DVC-COMSC/PythonPandas/main/png/iris.png"
urllib.request.urlretrieve(URL, "iris.png")

Image("iris.png")
````

#### How scikit-learn hands you the data
`datasets.load_iris()` returns an object, **not** a `DataFrame`:

| Attribute | What it is |
|---|---|
| `iris.data` | a 150 × 4 NumPy array — the measurements |
| `iris.target` | 150 numbers: `0`, `1`, or `2` — which species each flower is |
| `iris.feature_names` | the 4 column names |
| `iris.target_names` | `['setosa', 'versicolor', 'virginica']` — what `0/1/2` mean |

So building the `DataFrame` takes two steps: the measurements become the columns, then the species is added as one more column. **Run the next cell** — every problem in Sections 1–4 uses the `iris_df` it creates.

````
import pandas as pd
from sklearn import datasets

# Load the dataset (measurements + labels + documentation)
iris = datasets.load_iris()

# The 4 measurements become the columns
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the species as a 5th column. iris.target is 0/1/2; target_names turns
# those numbers into 'setosa' / 'versicolor' / 'virginica'.
iris_df['target'] = iris.target_names[iris.target]

iris_df
````

You should see a table of **150 rows × 5 columns** — the 4 measurements plus `target`. The row labels `0..149` on the left are the index; pandas made them for you.

Note the column names include the units: the column is `sepal length (cm)`, **not** `sepal_length`. You must type them exactly, parentheses and all.
